# 어느 구간에서 가장 많은 시간을 차지할까?
"status가 Enter이나 Change를 만나면 해당 아이디의 이전 alias기록을 갱신한다." 주석이 달린 부분임.

```python
def solution(record):
    import time
    
    t1=time.time()
    # record를 세 리스트(satus, id, alias)로 나누어 담기
    # Leave의 경우 alias가 없으므로 NA로 처리함.
    status=[]
    user_id=[]
    alias=[]
    issue_idx=[]
    for i in range(len(record)):
        if record[i].split(' ')[0]=='Leave':
            a,b=record[i].split(' ')
            c='NA'
            issue_idx.append(i)
        else:
            a,b,c=record[i].split(' ')
        status.append(a)
        user_id.append(b)
        alias.append(c)
        
    # NA를 별칭(alias)로 수정
    for idx in issue_idx:
        for i in range(len(record)):
            if idx<=i:
                break
            if user_id[i]==user_id[idx]:
                alias[idx]=alias[i]
    t2=time.time()
    print('#1',t2-t1)
    
    t3=time.time()
    # status가 Enter이나 Change를 만나면 해당 아이디의 이전 alias기록을 갱신한다.
    for i in range(len(record)):
        if status[i]=='Enter' or status[i]=='Change':
            for j in range(len(record)):
                if i<j:
                    break
                if user_id[i]==user_id[j]:
                    alias[j]=alias[i]
    t4=time.time()
    print('#2',t4-t3)
    
    t5=time.time()
    # Enter과 Leave기록을 문자열과 함께 리스트에 담아 출력한다.(Change는 기록남지 않는다)
    res=[]
    for i in range(len(record)):
        if status[i]=='Enter':
            res.append("%s님이 들어왔습니다." % alias[i])
        elif status[i]=='Leave':
            res.append("%s님이 나갔습니다." % alias[i])
    t6=time.time()
    print('#3',t6-t5)
    return res

if __name__=='__main__':
    solution(record)
```
