# 효율성 통과 못해서  데이터 입력구조를 바꿔봤음
## "세 개의 리스트변수"에서 "이중리스트" 구조로 
### 기존 : [Enter,Enter] [uid1234, uid4567] [Muzi,Prod]
### 변경 : [[Enter,uid1234,Muzi],[Enter,uid4567,Prodo]]
## 결과 : 더 느림! 기존 3개의 리스트변수가 더 빠름

### 새 방식

```python
def solution(record):
    import time
    a=time.time()
    info=[]
    issue_idx=[]
    for i in range(len(record)):
        info.append(list(record[i].split(' ')))
        if len(info[-1])==2:
            info[-1].append('NA')
            issue_idx.append(i)
   
    # NA를 별칭(alias)로 수정
    for idx in issue_idx:
        for i in range(len(record)):
            if idx<=i:
                break
            if info[idx][1]==info[i][1]:
                info[idx][2]=info[i][2]
    b=time.time()
    print(b-a)
    return
if __name__=='__main__':
    print(solution(record))
```

### 기존 방식

```python
def solution(record):
    import time
    e=time.time()
    # record를 세 리스트(satus, id, alias)로 나누어 담기
    # Leave의 경우 alias가 없으므로 NA로 처리함.
    status=[]
    user_id=[]
    alias=[]
    # Leave에 해당하는 alias를 저장하기 위해 해당 인덱스 번호를 저장함.
    issue_idx=[]
    for i in range(len(record)):
        if record[i].split(' ')[0]=='Leave':
            a,b=record[i].split(' ')
            c='NA'
            issue_idx.append(i)
        else:
            a,b,c=record[i].split(' ')
        status.append(a)
        user_id.append(b)
        alias.append(c)
    f=time.time()
    print(f-e)
    return
if __name__=='__main__':
    print(solution(record))
```
